```python
import torch
import tensor
import numpy as np
```

```python
X = np.random.randn(8, 16, 32)
X_tensor = tensor.Tensor.from_numpy(data=X, requires_grad=True)
X_torch = torch.from_numpy(X)
X_torch = X_torch.requires_grad_(True)
X_torch.is_leaf

y = X_tensor + 5
z = y + 10
m = z.mean()
```

```python
m.backward()
```

```python
y_torch = X_torch + 5
z_torch = y_torch + 10
m_torch = torch.mean(z_torch)
m_torch.backward()
```

```python
print(X_torch.grad)
```

```python
print(X_tensor.grad.data )
```

```python
np.allclose(X_tensor.grad.data, X_torch.grad.numpy())
```

```python
X_np = np.random.randn(8, 16, 32)
Y_np = np.random.randn(32, 24)
X_tens = tensor.Tensor.from_numpy(X_np, requires_grad=True)
Y_tens = tensor.Tensor.from_numpy(Y_np, requires_grad=True)

X_torch = torch.from_numpy(X_np)
X_torch.requires_grad_(True)
Y_torch = torch.from_numpy(Y_np)
Y_torch.requires_grad_(True)
```

```python
Z = X_tens.matmul(Y_tens)
W = Z.mean()
W.backward()
```

```python
Z_torch = X_torch.matmul(Y_torch)
W_torch = Z_torch.mean()
W_torch.backward()
```

```python
np.allclose(X_tens.grad.data, X_torch.grad.numpy()),np.allclose(Y_tens.grad.data, Y_torch.grad.numpy())
```
