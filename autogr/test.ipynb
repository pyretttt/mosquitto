```python
A = np.array([[1, 1, 1], [2, 2, 2], [5, 5, 5]])
B = np.array([[0, 1, 0], [1, 1, 0], [1, 1, 1]])
np.einsum('ij,jk->i', A, B)

res = np.empty((3,))

for i in range(3):
    total = 0
    for j in range(3):
        for k in range(3):
            total += A[i, j] * B[j, k]
    res[i] = total
res,np.einsum('ij,jk->ik', A, B)

```
