```python
import torch
```

```python
W = torch.randn((4, 3), requires_grad=False)
B = torch.randn((4, ), requires_grad=False)
C = torch.randn((4, ), requires_grad=True)

X = torch.ones((3, ), requires_grad=False)
```

```python
X_1 = torch.matmul(W, X) + B + C
X_1.requires_grad
```

```python
out = torch.sum(X_1) 
out.backward()
```
