```python
import torch
import tensor
import numpy as np
```

```python
X = np.random.randn(8, 16, 32)
X_tensor = tensor.Tensor.from_numpy(data=X, requires_grad=True)
X_torch = torch.from_numpy(X)
X_torch = X_torch.requires_grad_(True)
X_torch.is_leaf

y = X_tensor + 5
z = y + 10
m = z.mean()
```

```python
m.backward()
```

```python
y_torch = X_torch + 5
z_torch = y_torch + 10
m_torch = torch.mean(z_torch)
m_torch.backward()
```

```python
print(X_torch.grad)
```

```python
print(X_tensor.grad.data )
```

```python
np.allclose(X_tensor.grad.data, X_torch.grad.numpy())
```
