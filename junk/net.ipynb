```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cubic = open('06-cwnd-cubic.log', 'r').readlines()
reno = open('06-cwnd-reno.log', 'r').readlines()

cubic_cwnd = np.array([
    float(
        next(
            (kv for kv in line.split(' ') if kv.startswith('cwnd:'))
        )
        .split(":")[1]
    )
    for line in cubic
])

reno_cwnd = np.array([
    float(
        next(
            (kv for kv in line.split(' ') if kv.startswith('cwnd:'))
        )
        .split(":")[1]
    )
    for line in reno
])
```

```python
plt.bar(*np.unique(reno_cwnd, return_counts=True), label='Reno', color='skyblue', edgecolor='black', width=0.6)
plt.title("Reno")
```

```python
plt.bar(*np.unique(cubic_cwnd, return_counts=True), label='Cubic', color='skyblue', edgecolor='black', width=0.6)
plt.title("Cubic")
```
